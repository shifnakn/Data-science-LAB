```python
import numpy as np
```

```python

a = np.arange(10)
print(a)
```

```python
a = np.arange(25).reshape(5,5)
print(a)
print(a.shape)
```

```python
a = np.array([[1,2],[3,4]])
print(a)
```

```python
a = np.array([1,2,3])
print(type(a))
print(a.shape)
print(a[0],a[1],a[2])
a[0]=5
print(a)
```

```python
b=np.array([[1,2,3],[4,5,6]])
print(b.shape)
print(b[0,0],b[0,1],b[1,0])
print(b)
```
